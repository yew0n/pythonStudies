# #01. 필요한 패키지 및 샘플데이터 준비
## 1) 패키지 참조

````python
import numpy
from pandas import DataFrame
from pandas import read_csv
from matplotlib import pyplot
````

## 2) 데이터 가져오기
데이터출처 : KOSIS 국가통계포털

````python
교통사고csv = read_csv('http://itpaper.co.kr/demo/py/traffic.csv', encoding='euc-kr')
교통사고csv
````

## 3) 데이터 전처리 (년도별 합계 구하기)

````python
교통사고tmp = 교통사고csv.filter(['년도', '발생건수', '사망자수', '부상자수'])
교통사고df = 교통사고tmp.groupby(['년도']).sum()
교통사고df
````

# #02. 그래프에 대한 전역 환경 설정
이 코드가 다시 실행되기 전까지 모든 그래프 출력에 대해 적용된다.

````python
pyplot.rcParams['font.family'] = 'Malgun Gothic'
pyplot.rcParams['font.size'] = 16
pyplot.rcParams['figure.figsize'] = (20,10)   
````

# #03. 선 그래프
## 1) 인덱스가 숫자형식인 경우 특정 컬럼에 대한 선 그래프
DataFrame 객체의 plot() 함수를 사용한다.
### 기본옵션만 사용하기
<b>그래프 생성 시 주의사항</b> : 모든 코드가 하나의 블록 안에서 구현되어야 한다.

````python
교통사고df['발생건수'].plot(color='#ff0000')
pyplot.grid()
pyplot.legend()
pyplot.title('년도별 교통사고 발생현황')
pyplot.xlabel('년도')
pyplot.ylabel('교통사고 수')
pyplot.show()
````

### x 좌표에 적용할 텍스트의 리스트 생성

````python
xtext = list(교통사고df.index)
xtext
````

````python
for i, v in enumerate(xtext):
    xtext[i] = "%d년" % v
    
xtext
````

### x 축을 지정하여 그래프 표현하기 (DataFrame의 index가 숫자로만 되어있는 경우)
DataFrame의 index가 숫자 타입으로 구성되어 있는 경우 index값 자체가 x축의 좌표를 의미

x축의 좌표 어디에 어떤 값을 출력해야 한다는 의미로 `xticks()` 함수를 적용

<b>기존의 index와 이를 대체해 출력할 리스트를 xticks() 함수에 적용</b>



````python
교통사고df['발생건수'].plot(color='#ff0000')
pyplot.grid()
pyplot.legend()
pyplot.title('년도별 교통사고 발생현황')
pyplot.xlabel('년도')
pyplot.ylabel('교통사고 수')
pyplot.xticks(교통사고df.index, xtext)
pyplot.show()
````

## 2) 인덱스가 숫자형식이 아닌 경우의 선 그래프
### 인덱스가 문자열로 사용되도록 전처리
index가 문자열로 구성된 경우를 만들기 위해서 진행한 과정임!

````python
교통사고copy = 교통사고df.copy()

현재인덱스 = list(교통사고copy.index)

indexDict={}

for i,v in enumerate(xtext):
    before = 현재인덱스[i]
    indexDict[before] = v
    
print(indexDict)

교통사고copy.rename(index=indexDict, inplace=True)
교통사고copy
````

### x 축 좌표값으로 이루어진 배열(혹은 리스트)
인덱스가 문자열로 구성된 경우 그래프의 x축은 0부터 시작하는 좌표값으로 설정된다.

x축의 좌표 위치에 특정 텍스틀르 덧붙이기 위해서는 텍스트가 출력될 좌표값을 원소로 갖는 리스트(혹은 배열)을 미리 준비해야 한다.

````python
xpos = numpy.arange(len(교통사고copy.index))
xpos
````

### 구성된 좌표 리스트와 출력할 문자열 리스트를 활용하여 그래프 표시
DataFrame의 index가 숫자 타입으로 구성되어 있는 경우 index 값 자체가 x축의 좌표를 의미

x축의 좌표 어디에 어떤 값을 출력해야 한다는 의미로 xticks() 함수 적용

````python
교통사고copy['발생건수'].plot(color='#ff0000')
pyplot.grid()
pyplot.legend()
pyplot.title('년도별 교통사고 발생현황')
pyplot.xlabel('년도')
pyplot.ylabel('교통사고 수')
pyplot.xticks(xpos, xtext)
pyplot.show()
````

# #04. 세로막대 그래프
## 1) 특정 컬럼에 대한 시각화
### 기본옵션만 사용하기
```데이터프레임객체['컬럼'].plot.bar()```
- rot 파라미터 = x축 텍스트의 각도. 기본값 90
- width 파라미터 = 모든 막대가 공백없이 꽉 찬 상태를 1로 하여 비율적으로 지정. 기본값 0.5

막대그래프는 index의 데이터타입이 숫자, 문자열에 상관없이 x축이 0부터 시작되는 좌표계를 갖는다.

그러므로 xticks() 함수를 사용할 때 반드시 x축 좌표에 대한 지정이 이루어져야 한다.

````python
교통사고df['발생건수'].plot.bar(color='#8046eb', rot=0, width=0.7)
pyplot.grid()
pyplot.legend()
pyplot.title('년도별 교통사고 발생현황')
pyplot.xlabel('년도')
pyplot.ylabel('교통사고 수')
pyplot.xticks(xpos, xtext)
pyplot.show()
````

### 텍스트 출력 처리, y축 범위 조정
텍스트 출력은 그래프 상의 x, y 좌표를 구해 모든 문자열을 개별적으로 출력해야 한다.

이를 위해서 반복물이 사용된다.

````python
교통사고df['발생건수'].plot.bar(color='#8046eb', rot=0, width=0.7)
pyplot.grid()
pyplot.legend()
pyplot.title('년도별 교통사고 발생현황')
pyplot.xlabel('년도')
pyplot.ylabel('교통사고 수')
pyplot.xticks(xpos, xtext)

# y축의 범위설정 / x축의 범위를 설정할 경우는 xlim(min, max) 사용
pyplot.ylim(200000, 240000)

# 그래프에 텍스트 표시하기
# 그래프에 표시되는 실 데이터 = y축 좌표, x축 = 0부터 시작하는 좌표값을 가짐
for x, y in enumerate(list(교통사고df['발생건수'])):
    txt = "%s건" % format(y, ",")
    pyplot.text(x, y, txt, fontsize=14, color='#ff0000',
               horizontalalignment='center', verticalalignment='bottom')
    
pyplot.show()
````

# #06. 파이 그래프
데이터프레임의 특정 컬럼을 지정하여 전체를 기준으로 어느 정도의 비율을 차지하는지를 시각화하는 그래프
## 1) 파이그래프 기본형

````python
# 각 컬럼별로 평균 수하기
교통사고평균 = 교통사고df.mean()
교통사고평균
````

````python
# 구해진 결과를 사용해 데이터프레임 생성
교통사고평균df = DataFrame(교통사고평균, columns=['교통사고'])
교통사고평균df
````

````python
# 데이터프레임의 특정 컬럼에 대한 파이그래프 표시 -> 반드시 특정 컬럼을 지정해야 함
교통사고평균df['교통사고'].plot.pie()
pyplot.show()
````

## 2)파이그래프의 옵션 설정

````python
교통사고평균df['교통사고'].plot.pie(labels=['전체', '사망', '부상'],
                         colors=['#ff6600', '#0066ff', '#ffff00'],
                         explode=[0.0, 0.1, 0.0],
                         autopct='%0.1f%%',
                         shadow=True,
                         startangle=90)

# 제목지정
pyplot.title('교통사고 비율')

# y축 라벨 표시 안함
pyplot.ylabel(None)

pyplot.show()
````

## 3) 출력값에 대한 사용자 정의 함수 만들기
pie() 함수가 그래프의 조각 수 만큼 지정된 함수를 호출하며 각 조각이 차지하는 비율을 파라미터로 전달

지정된 함수는 파라미터로 전달되는 비율을 통해 몇 번째 항목의 데이터인지를 역으로 추적 -> 표시할 텍스트 구성 후 리턴

````python
def make_autopct(비율):
    print(비율)
    global 교통사고평균df
    합계 = 교통사고평균df['교통사고'].sum()
    수치 = int(round(비율*합계/100.0))      #비율과 총합을 사용해 수치값을 역산
    str = "%0.1f%%\n(%d건)" % (비율,수치)   #그래프에 각 섹션에 출력할 문자열 만들기
    return str

# 파이그래프의 각 섹션 수만큼 'make_autopct' 함수가 호출 -> 이 함수 안에서 각 섹션에 출력할 텍스트를 만들어 리턴
교통사고평균df['교통사고'].plot.pie(labels=None,
                          textprops={'color':'#ffffff', 'fontsize':18},
                         colors=['#ff6600', '#0066ff', '#669900'],
                         explode=[0.0, 0.1, 0.0],
                         autopct=make_autopct,
                         shadow=True,
                         startangle=90)
pyplot.ylabel(None)

# 범주 지정
# labels =각 섹션별 제목
# bbox_to_asnchor(x,y) = 0,0을 기준으로 범주의 위치(2차원 좌표)
pyplot.legend(labels=list(교통사고평균df.index), bbox_to_anchor=(0.95, 0.625))

pyplot.show()
````

# #07. 산점도 그래프
- 두 변수 간의 영향력을 보여주기 위해 가로축과 세로 축에 대한 <b>데이터 포인트</b>를 그리는데 사용
- <b>데이터 포인트</b>들이 뭉쳐있을수록 두 변수 간 관련성 정도가 높다고 해석

# 예) 교통사고 발생 건수와 부상자 수 간의 상관관계
- <b>첫 번째 변수</b> : 교통사고 발생건 수
- <b>두 번째 변수</b> : 교통사고 부상자 수

이 두 변수의 상관관계를 표현한 그래프를 산점도로 표현할 수 있다.

````python
# market -> o, v, ^, <, >, 8, s, p, *, h, H, D, d, P, X
교통사고csv.plot.scatter(x='발생건수', y='부상자수', color='#ff6600', marker='o')
pyplot.grid()
pyplot.title("교통사고 발생건수와 부상자 수의 상관관계")
pyplot.ylabel('부상자수')       # y축 라벨
pyplot.xlabel('발생건수')       # x축 라벨
pyplot.show()
````

## 산점도 그래프 해석
점들이 많이 모여있을수록 두 변수간 관계성이 높고, 점들이 흩어져 있을수록 두 변수간 관계성이 적다고 해석한다.
### 정의관계
x가 증가할 때 y도 증가하는 형태

````python
pyplot.rcParams['figure.figsize'] = (5,5)
pyplot.rcParams['font.family'] = 'Malgun Gothic'

df = DataFrame({'xdata':[1,2,3,4,5], 'ydata':[1,2,3,4,5]})
df.plot.scatter(x='xdata', y='ydata')
pyplot.show()
````

### 역의 관계
x가 증가할 때 y는 감소하는 형태

````python
df = DataFrame({'xdata':[1,2,3,4,5], 'ydata':[5,4,3,2,1]})
df.plot.scatter(x='xdata', y='ydata')
pyplot.show()
````

### 선형관계
점들이 직선에 가깝게 배치된 형태

<b>앞에서 제시한 두 샘플이 선형관계에 속한다</b>
### 비선형관계
점들이 직선이 안니 형태로 배치된 형태

````python
df = DataFrame({'xdata':[1,2,3,4,5], 'ydata':[3,7,5,9,1]})
df.plot.scatter(x='xdata', y='ydata')
pyplot.show()
````

## 상관분석
연속형 변수로 측정된 두 변수 간의 선형적 관계를 분석하는 기법

연속형 변수는 산술 평균을 계산할 수 있는 숫자형의 데이터이며, 선형적 관계라 함은 흔히 비례식이 성립되는 관계를 말한다.

    예를 들어 A변수가 증가함에 따라 B변수도 증가되는지/감소하는지를 분석하는 것

상관분석은 두 변수간 선형적 관계에 초점을 두고 있으며 아래와 같은 질문을 던지며 결과 해석을 수행한다.

- <b>선형관계를 갖는가?</b>
- <b>어느 방향으로 형성되어 있는가?</b>
- <b>그 관계의 정도는 얼마나 큰가?</b>


##  상관계수
상관분석에서 두 변수 사이의 선형적인 관계 정도를 나타내기 위해 사용되는 값

    수학적 공식에 의해 구해야 한다.
    
상관분석에는 측정데이터에 따라 여러가지 분석방법(피어슨 상관분석/스퍼만 상관분석 등)이 있으나, 일반적으로 피어슨 상관계수를 의미한다.

상관계수 | 설명 |
---- | ---- |
1 | 직선형 정의관계 |
-1 | 직선형 역의관계 |
0 | 모든 점들이 분산되어 있는 형태 |

> 상관계수 1과 0 사이에 소수점 형태로 점의 분포도를 표현한다.

## 추세선
차트 위에 그려진 대각선들.<br>
이는 특정 데이터 지점을 연결하며 분석가가 보다 쉽게 데이터의 움직임을 시각화하고 추세를 알 수 있게 한다.<br>
추세선은 기술적 분석의 가장 기초적 도구 중 하나로 간주된다.

### numpy를 활용한 추세선 방정식 구하기

````python
x = 교통사고csv['발생건수']
y = 교통사고csv['부상자수']

# 추세선을 위한 계산 - 1차원의 pollynomial(다항식)을 계산하기 위한 코드
z = numpy.polyfit(x,y,1) # (X, Y, 차원) 정의
f = numpy.poly1d(z)      # 1차원 다항식에 대한 연산을 캡슐화
f
````

### 추세선을 포함한 산점도 그래프

````python
pyplot.rcParams['font.family'] = 'Malgun Gothic'
pyplot.rcParams['font.size'] = 16
pyplot.rcParams['figure.figsize'] = (20,10) 

교통사고csv.plot.scatter(x='발생건수', y='부상자수', color='#ff6600', marker='o')
pyplot.grid()
pyplot.title("교통사고 발생건수와 부상자 수의 상관관계")
pyplot.ylabel('부상자수')       # y축 라벨
pyplot.xlabel('발생건수')       # x축 라벨
pyplot.plot(x, f(x), '#ff0000')
pyplot.show()
````
