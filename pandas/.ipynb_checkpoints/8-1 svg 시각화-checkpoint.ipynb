# #02. 지도 가져오기
위키미디어에서 <b>Seoul districts.svg</b> 키워드로 검색하여 서울 지도 이미지를 내려받아 <b>map_seoul.svg</b>라는 이름으로 작업폴더에 추가
> https://commons.wikimedia.org/wiki/File:Seoul_districts.svg?uselang=ko<br>
        http://itpaper.co.kr/demo/py/map_seoul.svg

```python
map_svg = None
with open('./map_seoul.svg', 'r') as f:
    map_svg = f.read()
    
#print(type(map_svg))
#map_svg
```

## 2) 원하는 내용만 가져오기
svg 파일의 내용을 BeautifulSoup 객체로 변환 후 id 속성을 갖는 path 태그만 추출
> lxml 관련 경고가 표시될 경우 무시

```python
soup = BeautifulSoup(map_svg)
paths = soup.select('path[id]')

#출력결과는 path 단위를 저장한 리스트
# path 하나당 지역 한 곳 -> 리스트의 원소 하나 = 지역 하나
paths
```

# #03. 데이터 시각화
## 1) 시각화 준비
단계는 분석가가 임의로 설정, 색상값을 1단계~높은단계 순으로 점점 진하게 되도록 구성

```python
# 사용할 색상값 (단계별 6개 색상 준비)
colors = ['#F1EEF6', '#D4B9DA', '#C994C7', '#DF65B9', '#DD1C77', '#980043']
```

## 2) 지도의 각 구역만큼 반복

```python
for p in paths:
    #print(p['id'])
    
    # 데이터프레임에서 지역명에 대한 복지시설 수 조회하기
    tmp_df = df.query("COUNTY == '"+p['id']+"'")
    #print(tmp_df)
    
    
    # 복지시설 수
    count = tmp_df.loc[tmp_df.index[0], 'NUMBER']
    #print("%s >> %d" % (p['id'], count))
    
     # 복지시설 수에 따라 단계값 설정 (단계는 색상의 수에 따름)
    if count>250 : color_index=5
    elif count>200 : color_index=4
    elif count>150 : color_index=3
    elif count>100 : color_index=2
    elif count>50 : color_index=1
    else : color_index=0
        
    # 가져온 svg 이미지의 path 태그들의 fill 속성에 선정한 단계의 색상값 적용
    p['fill'] = colors[color_index]
    
    print("%s >> %d, color_index=%d, color=%s, fill=%s" % (p['id'], count, color_index, colors[color_index], p['fill']))
```

## 3) 재구성된 내용을 추출하여 새로운 scg 소스코드 얻기

```python
# bs4 객체의 내용을 문자열로 리턴
new_seoul_svg = soup.prettify()
new_seoul_svg
```

### 파일로 저장

```python
# 저장된 파일은 윈도우 폴더창에서 직접 더블클릭 후 웹 브라우저를 통해 확인
with open('new_seoul_svg.svg', 'w', encoding='utf-8') as f:
    f.write(new_seoul_svg)
    
# jupyter에서 svg 이미지 표시하기
# 사용방법 -> SVG(소스문자열) 혹은 SVG(파일경로)
SVG('new_seoul_svg.svg')
```

# #04. 전국의 학교 분포 시각화
## 1) 데이터 수집

```python
# 엑셀파일을 원격으로 내려받아 데이터프레임으로 변환 (다소 시간 소요)
학교df = read_excel('http://itpaper.co.kr/demo/py/school2019.xlsx', engine='openpyxl')
학교df
```

## 2) 데이터 전처리 및 정제
### 필요한 필드만 추출하기

```python
학교추출df = 학교df.filter(['학교명', '학교급구분', '소재지도로명주소'])
학교추출df
```

### 데이터 정제
각 열별로 결측치 확인

```python
학교추출df.isna().sum()
```

결측치가 존재하므로 해당 데이터 삭제

```python
최종df = 학교추출df.copy()
최종df.dropna(inplace=True)
최종df.isna().sum()
```

### 주소에서 시, 도 이름만 추출하기
주소값에서 첫 번째 공백의 위치를 별도 컬럼으로 추가

```python
최종df['idx'] = 최종df['소재지도로명주소'].str.find(' ')
최종df
```

### 주소에서 첫 번째 공백의 위치까지 잘라낸 값을 새로운 컬럼으로 추가
전체 반복문 처리이므로 시간 소요

```python
결과df = 최종df.copy()

for i in 결과df.index:
    addr = 결과df.loc[i, '소재지도로명주소']
    idx = 결과df.loc[i, 'idx']
    city = addr[:idx]
    결과df.loc[i, '시도'] = city
    
결과df
```

### 시, 도의 데이터 분포 확인
<b>전북, 전남, 충북, 충남, 대구</b>의 이름에 대한 이상치가 존재하므로 보정 필요

```python
결과df['시도'].value_counts()
```

### 시, 도 이름 보정

```python
값보정df = 결과df.copy()

# 보정해야 할 이름 조건 나열
conditions = [ (값보정df['시도']=='경남'),
              (값보정df['시도']=='경북'),
              (값보정df['시도']=='전남'),
              (값보정df['시도']=='전북'),
              (값보정df['시도']=='충남'),
              (값보정df['시도']=='충북'),
              (값보정df['시도']=='대구') ]

#보정될 이름값 나열
name = ['경상남도', '경상북도', '전라남도', '전라북도', '충청남도', '충청북도', '대구광역시']

# 시도 이름 보정
# -> default는 모든 조건이 맞지 않을 경우 적용할 값
# -> durltjsms `시도` 컬럼의 값을 그대로 사용하도록 지정
값보정df['시도'] = numpy.select(conditions, name, default = 값보정df['시도'])

값보정df
```

## 3) 데이터 분석 - 지역별, 학교별 집계
### 사용할 컬럼 추출

```python
결과추출df = 값보정df.filter(['시도', '학교급구분', '학교명'])
결과추출df
```

### 그룹분석
그룹분석 결과를 피벗테이블로 재구성하기 위해 `as_index=False` 파라미터를 사용하여 인덱스를 병합하지 않도록 지정한다.

```python
집계df = 결과추출df.groupby(['시도', '학교급구분'], as_index=False).count()
집계df
```

### 컬럼이름 변경

```python
집계결과df = 집계df.rename(columns={'학교명':'학교수'})
집계결과df
```

### 집계 결과를 피벗 테이블로 재배치

```python
pivot_df = 집계결과df.pivot('시도', '학교급구분', '학교수')
pivot_df
```

## 4) 데이터 시각화
### SVG 파일 열기
http://itpaper.co.kr/demo/py/map_korea.svg 파일을 내려받아 사용하세요.

```python
map_svg = None
with open('./map_korea.svg', 'r', encoding='utf-8') as f:
    map_svg = f.read()
    
#SVG(map_svg)
```

### 지도 시각화 준비
사용할 색상값 (단계별로 6개 색상 준비)

```python
# 사용할 색상값 (단계별 6개 색상 준비)
colors = ['#F1EEF6', '#D4B9DA', '#C994C7', '#DF65B9', '#DD1C77', '#980043']
```

svg 데이터 로드

```python
soup = BeautifulSoup(map_svg)
#soup
```

id 속성을 갖는 `<g>` 태그와 `<path>` 태그 추출
> `<g>` = <b>시도</b> `<path>` = <b>구,군</b>을 의미하는 `<g>`의 하위요소

```python
glist = soup.select('svg>g[id], svg>path[id]')
#glist
```

### 초등학교 데이터만 시각화하기

```python
for item in glist:
    #print(item['id'])
    
    # 지도의 id값(시도이름)이 dataframe의 index에 포함되어 있지 않다면 다음 조건으로 이동
    if item['id'] not in pivot_df.index:
        continue
    
    # 각 지역별 학교 수
    count = pivot_df.loc[item['id'], '초등학교']
    #print(count)
    
    # 학교 수에 따라 단계값 설정 (단계는 색상의 수에 따름)
    if count>500 : color_index=5
    elif count>400 : color_index=4
    elif count>300 : color_index=3
    elif count>200 : color_index=2
    elif count>100 : color_index=1
    else : color_index=0
        
    # 각 item별 fill 속성 확인
    #print(item.attrs)
    
    # item에 fill 속성이 있는 경우 -> 세종특별자치시
    if 'fill' in item.attrs:
        item['fill']=colors[color_index]
    #그렇지 않은 경우 -> 나머지 지역
    else:
        #현재요소(item)의 하위 항목들에 색상 적용
        for p in item.select('path'):
            if 'fill' in p.attrs:
                p['fill'] = colors[color_index]
                
초등학교svg = soup.prettify()

with open('1.svg', 'w', encoding='utf-8') as f:
    f.write(초등학교svg)
    
SVG('1.svg')
```
