# 분석용 데이터 구축 실습 - 신예원

## 1) 문항1 - 데이터프레임 추출 및 결측치 확인+제거

```python
# TAG로부터 원하는 내용을 추출하는 클래스 -> SVG 이미지의 핸들링을 위함
from bs4 import BeautifulSoup

from pandas import DataFrame
from pandas import read_excel
from matplotlib import pyplot
```

### 데이터프레임 추출

```python
# 엑셀파일을 원격으로 내려받아 데이터프레임으로 변환 (다소 시간 소요)
학교df = read_excel('http://itpaper.co.kr/demo/py/school2019.xlsx', engine='openpyxl')
학교df
```

### '학교명', '학교급구분', '소재지도로명주소' 데이터프레임 추출

```python
학교추출df = 학교df.filter(['학교명', '학교급구분', '소재지도로명주소'])
학교추출df
```

### 데이터 정제 (결측치 존재 확인 및 소거)

```python
학교추출df.isna().sum()
```

```python
최종df = 학교추출df.copy()
최종df.dropna(inplace=True)
최종df.isna().sum()
```

## 2) 문항 2 - 데이터 변환 적재하기
### '시도', '학교급구분', '학교명' 데이터프레임 생성

```python
최종df['idx'] = 최종df['소재지도로명주소'].str.find(' ')
최종df

결과df = 최종df.copy()

for i in 결과df.index:
    addr = 결과df.loc[i, '소재지도로명주소']
    idx = 결과df.loc[i, 'idx']
    city = addr[:idx]
    결과df.loc[i, '시도'] = city
    
결과df = 결과df.drop(columns=['idx', '소재지도로명주소'])
결과df
```

### 각 시도에 따른 학교급 구분 별 학교 수 확인

```python
# 각 시도 명 중 보정이 필요한 값에 대한 값보정 수행
값보정df = 결과df.copy()

# 보정해야 할 이름 조건 나열
conditions = [ (값보정df['시도']=='경남'),
              (값보정df['시도']=='경북'),
              (값보정df['시도']=='전남'),
              (값보정df['시도']=='전북'),
              (값보정df['시도']=='충남'),
              (값보정df['시도']=='충북'),
              (값보정df['시도']=='대구') ]

#보정될 이름값 나열
name = ['경상남도', '경상북도', '전라남도', '전라북도', '충청남도', '충청북도', '대구광역시']

# 시도 이름 보정
# -> default는 모든 조건이 맞지 않을 경우 적용할 값
# -> durltjsms `시도` 컬럼의 값을 그대로 사용하도록 지정
값보정df['시도'] = numpy.select(conditions, name, default = 값보정df['시도'])

값보정df
```

#### 학교급구분 별 몇 개의 학교가 있는지 확인

```python
#시도별/학교급구분에 따라 갯수 count
집계df = 값보정df.groupby(['시도', '학교급구분'], as_index=False).count()
#집계df

# 학교명 컬럼이름을 학교수로 CHANGE
집계결과df = 집계df.rename(columns={'학교명':'학교수'})
집계결과df
```

## 3) 문항3 - 문항2의 데이터프레임 결과 확인을 위해 피벗테이블로 데이터 재구성

```python
pivot_df = 집계결과df.pivot('시도', '학교급구분', '학교수')
pivot_df
```

## 4) 문항4 - 지역별 학교 분포를 시각화하여 데이터 검증
### 그래프 출력 기본 설정

```python
pyplot.rcParams['font.family'] = 'Malgun Gothic'
pyplot.rcParams['font.size'] = 16
pyplot.rcParams['figure.figsize'] = (20,10)   
```

### 지역별 학교분포 시각화

```python
pivot_df.plot.bar(rot=45, width=0.7)
pyplot.grid()

pyplot.legend()
pyplot.legend(bbox_to_anchor=(1,1))

pyplot.title('지역별 초,중,고등학교 분포 현황')
pyplot.xlabel('시도')
pyplot.ylabel('학교수')

pyplot.show()
```
